```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report


df_banknote = pd.read_csv('data_banknote_authentication.csv')


X_bank = df_banknote.drop(columns=['class'])
y_bank = df_banknote['class']


X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(
    X_bank, y_bank, test_size=0.20, random_state=20, stratify=y_bank
)


svc_linear = SVC(kernel='linear')
svc_linear.fit(X_train_b, y_train_b)

y_pred_linear = svc_linear.predict(X_test_b)
cm_linear = confusion_matrix(y_test_b, y_pred_linear)
cr_linear = classification_report(y_test_b, y_pred_linear)


svc_rbf = SVC(kernel='rbf')
svc_rbf.fit(X_train_b, y_train_b)


y_pred_rbf = svc_rbf.predict(X_test_b)
cm_rbf = confusion_matrix(y_test_b, y_pred_rbf)
cr_rbf = classification_report(y_test_b, y_pred_rbf)


print("=== LINEAR KERNEL SVM ===")
print("Confusion Matrix:\n", cm_linear)
print("\nClassification Report:\n", cr_linear)

print("\n=== RBF KERNEL SVM ===")
print("Confusion Matrix:\n", cm_rbf)
print("\nClassification Report:\n", cr_rbf)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score


df_hw = pd.read_csv('weight-height (1).csv')


df_hw['Height_cm'] = df_hw['Height'] * 2.54
df_hw['Weight_kg'] = df_hw['Weight'] * 0.45359237


X_hw = df_hw[['Height_cm']]
y_hw = df_hw['Weight_kg']


X_train_h, X_test_h, y_train_h, y_test_h = train_test_split(
    X_hw, y_hw, test_size=0.20, random_state=42
)


scaler_minmax = MinMaxScaler()
X_train_norm = scaler_minmax.fit_transform(X_train_h)
X_test_norm = scaler_minmax.transform(X_test_h)


scaler_std = StandardScaler()
X_train_std = scaler_std.fit_transform(X_train_h)
X_test_std = scaler_std.transform(X_test_h)


knn_unscaled = KNeighborsRegressor(n_neighbors=5)
knn_unscaled.fit(X_train_h, y_train_h)
y_pred_unscaled = knn_unscaled.predict(X_test_h)
r2_unscaled = r2_score(y_test_h, y_pred_unscaled)


knn_norm = KNeighborsRegressor(n_neighbors=5)
knn_norm.fit(X_train_norm, y_train_h)
y_pred_norm = knn_norm.predict(X_test_norm)
r2_norm = r2_score(y_test_h, y_pred_norm)


knn_std = KNeighborsRegressor(n_neighbors=5)
knn_std.fit(X_train_std, y_train_h)
y_pred_std = knn_std.predict(X_test_std)
r2_std = r2_score(y_test_h, y_pred_std)

# Print R^2 values
print(f"R^2 Unscaled:     {r2_unscaled:.6f}")
print(f"R^2 Normalized:   {r2_norm:.6f}")
print(f"R^2 Standardized: {r2_std:.6f}")
```
